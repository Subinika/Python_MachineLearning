```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

data = [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12,100,108]
print(data)
```

```python
data = sorted(data)
print(data)
```

```python
sns.histplot(data=data , kde=True)
plt.show()
```

```python
def detect_outlier(data , threshold):
    mean = np.mean(data)
    standard_deviation = np.std(data)
    z_scores = np.abs((data - mean) / standard_deviation)

    return z_scores > threshold
    
```

```python
outliers  = detect_outlier(data ,1)
print(outliers)

true_indices = []
outlier_value = []
for i in range(len(outliers)):
    if outliers[i]:
        true_indices.append(i)

print("Index of True value is:", true_indices)

```

```python
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19,
           107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15,
           12, 10, 14, 13, 15, 10]
```

```python
# Step 1: sort the dataset in ascending order
dataset = sorted(dataset)
print(dataset)
```

```python
# Step 2: calculate Q1 and Q3
# Q1: 25% percentile
# Q3: 75% percentile

Q1, Q3 = np.percentile(dataset, [25, 75])

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
```

```python
# Step 3: Compute IQR

iqr = Q3 - Q1

print(f"IQR: {iqr}")
```

```python
# Step 4: Compute Lower Fence and Upper Fence

lower_fence = Q1 - (1.5 * iqr)
upper_fence = Q3 + (1.5 * iqr)

print(f"Lower Fence: {lower_fence}")
print(f"Upper Fence: {upper_fence}")
```

```python
# Step 5: Make Conclusion

def detect_outlier_using_iqr(data, iqr, lower_fence, upper_fence):
  outliers = []
  for i in data:
    if i < lower_fence or i > upper_fence:
      outliers.append(i)

  return outliers
```

```python
# function call
outliers = detect_outlier_using_iqr(dataset, iqr, lower_fence, upper_fence)
print(f'Detected Outliers:\n {outliers}')
```
