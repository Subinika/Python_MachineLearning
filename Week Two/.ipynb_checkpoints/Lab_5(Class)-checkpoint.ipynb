```python
# 1. Create an abstract class called `Animal` with abstract methods:
from abc import ABC , abstractmethod

class Animal(ABC):

    def __init__(self ,name, sound , movement , food ):
        self.sound = sound
        self.name = name 
        self.movement = movement
        self.food = food

    @abstractmethod
    def speak(self):
        pass

    @abstractmethod
    def move(self):
        pass

    @abstractmethod
    def eat(self):
        pass
        
```

```python
# 2. Implement three concrete classes: `Mammal`, `Bird`, and `Fish`, inheriting from the `Animal` class. Implement the abstract methods accordingly.

class Mammal(Animal):
    def __init__(self, name , sound, movement ,food):
        super().__init__(name,sound, movement ,food)

    def speak(self):
        print(f"{self.name} make {self.sound}. ")

    def move(self):
        print(f"{self.name} can {self.movement}. ")

    def eat(self):
        print(f"{self.name} eat {self.food}. ")


mammal = Mammal("Mammals","sound" , "walk" , "food")
mammal.speak()
mammal.move()
mammal.eat()

print("\n")
class Bird(Animal):
    def __init__(self ,name, sound , movement , food):
        super().__init__(name,sound, movement , food)


    def speak(self):
        print(f"{self.name} can make {self.sound}")

    def move(self):
        print(f"{self.name} can  {self.movement}")

    def eat(self):
        print(f"{self.name} eat {self.food}")

birds = Bird("Bird","sound" , "fly","insect")
birds.speak()
birds.move()
birds.eat()

print("\n")
class Fish(Animal):
    def __init__(self ,name, sound , movement , food):
        super().__init__(name,sound, movement , food)


    def speak(self):
        print(f"{self.name} can make {self.sound}")

    def move(self):
        print(f"{self.name} can  {self.movement}")

    def eat(self):
        print(f"{self.name} eat {self.food}")

fish = Fish("Fish","sound" , "swim","insect")
fish.speak()
fish.move()
fish.eat()
```

```python
# 3. Create concrete classes for specific animals within each category:
#  - For `Mammal`: Implement classes like `Dog` and `Cat`.
#   - For `Bird`: Implement classes like `Eagle` and `Penguin`.
#   - For `Fish`: Implement classes like `Salmon` and `Goldfish`.
class Dog(Mammal):
    def __init__(self, sound):
        super().__init__("Dog","bark", "run", "dog food")
        self.sound = sound

    def bark(self):
        print(f"The Dog {self.sound} when see unknown")


class Cat(Mammal):
    def __init__(self):
        super().__init__("Cat","meow","jump","fish")

class Eagle(Bird):
    def __init__(self, high):
        self.high = high
        super().__init__("Eagle","chow","fly","snake")

    def fly(self):
        print(f"Eagle {self.high} high in sky ")

class Penguin(Bird):
    def __init__(self):
        super().__init__("Penguin","quic","walk","fish")

class Salmon(Fish):
    def __init__(self, under_water):
        self.under_water = under_water
        super().__init__("Salmon","jii","swim","insect")

    def swim(self):
        print(f"Salmon swmin {self.under_water} under water")

class GoldFish(Fish):
    def __init__(self):
        super().__init__("GoldFish","iij","swim","insect")
cat = Cat()
dog = Dog("bark")
eagle = Eagle("fly")
penguin = Penguin()
salmon = Salmon("swim")
goldFish = GoldFish()


dog.speak()
dog.eat()
dog.bark()
dog.move()
print("\n")
cat.speak()
cat.eat()
cat.move()
print("\n")
eagle.speak()
eagle.eat()
eagle.fly()
eagle.move()
print("\n")
penguin.speak()
penguin.eat()
penguin.move()
print("\n")
salmon.speak()
salmon.eat()
salmon.swim()
salmon.move()
print("\n")
goldFish.speak()
goldFish.eat()
goldFish.move()
```

```python
#1. Create a base class called `Vehicle` with the following attributes:
#   - `make`: Make of the vehicle (e.g., Ford, Honda).
#   - `model`: Model of the vehicle (e.g., Civic, F-150).
#   - `year`: Year of manufacture.
#   - `fuel_type`: Type of fuel the vehicle uses (e.g., Gasoline, Electric).
class Vehicle:

    def __init__(self, make , model_number , mfd_year , type):
        self.make = make
        self.model_number = model_number
        self.mfd_year = mfd_year
        self.type = type

    def make(self):
        print(f"Make - {self.maker}")

    def model(self):
        print(f"Model - {self.model_number}")

    def year(self):
        print(f"Manufactured Year - {self.mfd_year}")

    def fuel_type(self):
        print(f"Fuel Type - {self.type}")

    def call(self):
        self.year()
        self.model()
        self.make()
        self.fuel_type()

class Car(Vehicle):
    def __init__(self, maker , model_number , mfd_year , type):
        super().__init__(maker , model_number , mfd_year , type )

class ElectricCar(Car):
    def __init__(self,  maker, model_number, mfd_year, type, battery_capacity):
        self.battery_capacity = battery_capacity
        super().__init__( maker, model_number, mfd_year, type)

    def battery(self):
        print(f"Battery Capacity - {self.battery_capacity}")

eCar = ElectricCar("Maruti" , "F-50" , "2023 April" , "Fuel-Petrol" ,"120 KW")
eCar.call()
eCar.battery()

print("\n")

car = Car("Suzuki", "C-20", "2023 jan", "Fuel")
car.call()

print("\n")

class Truck(Vehicle):
    def __init__(self, maker , model_number , mfd_year , type):
        super().__init__( maker , model_number , mfd_year , type)

class HybridTruck(Car):
    def __init__(self,  maker, model_number, mfd_year, type, motor_capacity):
        self.motor_capacity = motor_capacity
        super().__init__( maker, model_number, mfd_year, type)
    def electric_motor(self):
        print(f"Electric Motor Power - {self.motor_capacity}")

hTruck = HybridTruck("Mahendra" , "T-50" , "2021 April" , "Electric" ,"20000 Rpm")
hTruck.call()
hTruck.electric_motor()

print("\n")

truck = Truck("TATA", "T-60", "2021 april","Fuel")
truck.call()
```
