```python
# polymorphism
# Method Overriding
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Example of polymorphism
def make_animal_speak(animal):
    animal.speak()

dog = Dog()
cat = Cat()

make_animal_speak(dog)  # Calls Dog's speak method
make_animal_speak(cat)  # Calls Cat's speak method
```

```python
"""
Q. Write a python program to print circumference of circle.
  - `create class attributes to initialize pi=3.14`
  - `create __init__() method to initialize radius of circle for each instance`
  - `create method named get_circumference() to return circumference of a circle`
  - `create method named set_radius() to reset the radius of circle.`
"""
class Circle:
    # Class attribute to initialize pi
    pi = 3.14

    def __init__(self, radius):
        # Initializing radius for each instance
        self.radius = radius

    def get_circumference(self):
        # Method to return circumference of a circle
        circumference = 2 * Circle.pi * self.radius
        return circumference

    def set_radius(self, new_radius):
        # Method to reset the radius of the circle
        self.radius = new_radius

# Creating an instance of the Circle class with radius 5
circle_instance = Circle(5)

# Getting the circumference of the circle
circumference_value = circle_instance.get_circumference()
print(f"Circumference of the circle with radius 5: {circumference_value}")

# Setting a new radius for the circle
circle_instance.set_radius(7)

# Getting the circumference again with the updated radius
new_circumference_value = circle_instance.get_circumference()
print(f"Circumference of the circle with updated radius 7: {new_circumference_value}")
```

```python
# Inheritance 
# Single Inheritance 
#  Write a python program to create a base class named Vehicle with public object attributes (name, color) and public object methods (start, stop). From  base class create a 2 derived class named Car having object attributes (fuel_type), and object methods (drive) and Bike having attrbutes (engine_capacity), methods (ride)
class Vehicle:
    def __init__(self,name,color):
        self.name = name
        self.color = color

    def start(self):
        print(f"The {self.color} {self.name} is starting.")

    def stop(self):
        print(f"The {self.color} {self.name} is stopping.")


class Car(Vehicle):
    def __init__(self, name, color, fuel_type):
        super().__init__(name,color)
        self.fuel_type = fuel_type

    def drive(self):
        print(f"The {self.color} {self.name} is driving.")
        
car = Car("Car","Red","Petrol")
car.start()
car.drive()
car.stop()

class Bike(Vehicle):
    def __init__(self, name, color, engine_capacity):
     super().__init__(name, color) ## super is used to call base class constructor
     self.engine_capacity = engine_capacity

    def ride(self):
     print(f"The {self.color} {self.name} is being ridden.")


car = Car("Car", "Red", "Petrol")

bike = Bike("Motorcycle", "Blue", 250)

print(car.name) 
print(bike.color)

car.start() 
bike.stop()

car.drive() 
bike.ride()
```

```python
# Multiple Inheritance
# Write a python program to create 2 Base class named Wheels and Engine. From the 2 base classes, create a derived class named Car using the concept of multiple inheritance
# Base class 1
class Engine:
    def start(self):
        print("The engine is starting.")

    def stop(self):
        print("The engine has stopped.")

# Base class 2
class Wheels:
    def rotate(self):
        print("The wheels are rotating.")

# Derived class
class Car(Engine, Wheels):
  def drive(self):
    print('The Car is being driven')

# Create a instance  of Car
car = Car()
print(type(car))

# access methods of base class "Engine"
car.start() 
car.stop()

# access methods of base class "Wheel"
car.rotate() # output: The wheels are rotating.
```

```python
# Multilevel Inheritance 
# Base class
class Vehicle:
  def __init__(self, name):
    print("I am from Init")
    self.name = name

  def start(self):
    print(f"The {self.name} is starting.")

# Derived Class inheriting from Vehicle
class Car(Vehicle):
  def __init__(self, name):
    Vehicle.__init__(self, name)
  def drive(self):
    print(f"The {self.name} is being driven.")

# Derived Class inheriting from Car
class SportsCar(Car):
  def __init__(self, name):
    Car.__init__(self, name)
  def accelerate(self):
    print(f"The {self.name} is accelerating.")

# Create instance of sportscar
sports_car = SportsCar("sports car")

# access class "Vehicle" method
sports_car.start()  

# access class "Car" method
sports_car.drive() 

# access class "SportsCar"  method
sports_car.accelerate()
```

```python
# Hierarchical inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def bark(self):
        print(f"{self.name} barks")

class Cat(Animal):
    def meow(self):
        print(f"{self.name} meows")

class Cow(Animal):
    def moo(self):
        print(f"{self.name} moos")

# Create instances of the derived classes
dog = Dog("Buddy")
cat = Cat("Whiskers")
cow = Cow("MooMoo")

# Accessing common method from the base class
dog.speak()  # Output: Buddy makes a sound
cat.speak()  # Output: Whiskers makes a sound
cow.speak()  # Output: MooMoo makes a sound

# Accessing specific methods from the derived classes
dog.bark()   # Output: Buddy barks
cat.meow()   # Output: Whiskers meows
cow.moo()    # Output: MooMoo moos
```

```python
#Polymorphism
## write your program here
class Car:
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def vehicle_type(self):
    print(f'{self.name} {self.color} is Electric type')

class Truck:
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def vehicle_type(self):
    print(f'{self.name} {self.color} is Diesel type')

class Bike:
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def vehicle_type(self):
    print(f'{self.name} {self.color} is Petrol type')


# Instantiate object for 3 classes
car = Car('car', 'red')
truck = Truck('truck', 'black')
bike = Bike('bike', 'green')

car.vehicle_type()
truck.vehicle_type()
bike.vehicle_type()

# Method Invoke
for obj in (car, truck, bike):
  obj.vehicle_type()
  print("---------------")
    
```
